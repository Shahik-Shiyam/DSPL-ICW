```python
import pandas as pd

df = "/content/UNdata_Export_20250422_214745039.csv"

pd.read_csv(df)
```

```python
#data preprocessing


df = df.drop(columns=['OID', 'Magnitude', 'Country or Area'])

df.head()
```

```python
#Data exploration
dtype = df.dtypes
print(dtype)
```

```python
duplicate_rows = df[df.duplicated()]

# Print the number of duplicate rows
print(f"Number of duplicate rows: {len(duplicate_rows)}")
```

```python
null_counts = df.isnull().sum()

# Print the null counts for each column
print("Null Value Counts per Column:")
print(null_counts)
```

```python
total_nulls = df.isnull().sum().sum()
print(f"\nTotal Null Values in DataFrame: {total_nulls}")
```

```python
df.to_csv("Interest_Rates.csv", index=False)
```

```python
#Feature Enginerring
# Segmentation by Risk Level
risk_levels = {
    "Treasury Bill Rate": "Very Low",
    "Advance Rate (Central Bank Lending)": "Very Low",
    "Interbank Call Money Rate": "Moderate",
    "Commercial Paper Rate": "Moderate to High",
    "Certificate of Deposit (CD) Rate": "Low to Moderate",
    "Bankers' Acceptance Rate": "Low to Moderate"
}

print("Segmentation by Risk Level:\n")
for rate, risk in risk_levels.items():
    print(f"- {rate}: {risk}")

# --------------------------------------------

# Segmentation by Government or Private Sector
sector_types = {
    "Treasury Bill Rate": "Government",
    "Advance Rate (Central Bank Lending)": "Government",
    "Interbank Call Money Rate": "Private",
    "Commercial Paper Rate": "Private",
    "Certificate of Deposit (CD) Rate": "Private",
    "Bankers' Acceptance Rate": "Private"
}

print("\nSegmentation by Government or Private Sector:\n")
for rate, sector in sector_types.items():
    print(f"- {rate}: {sector}")
```
